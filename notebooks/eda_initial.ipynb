```python
import pandas as pd

# Load dataset
df = pd.read_csv('../data/raw/german_credit_data.csv')
```

```python
def risk_label(row):
    # Custom scoring logic (simple rule-based)
    score = 0
    
    if row['Credit amount'] > 5000:
        score += 1
    if row['Duration'] > 24:
        score += 1
    if pd.isna(row['Checking account']) or row['Checking account'] in ['none', 'little']:
        score += 1
    if pd.isna(row['Saving accounts']) or row['Saving accounts'] in ['little', 'moderate']:
        score += 1
    
    return 'Bad' if score >= 2 else 'Good'

# Apply to each row
df['Risk'] = df.apply(risk_label, axis=1)

# Check class distribution
print(df['Risk'].value_counts())
```

```python
df.to_csv('../data/processed/credit_data_with_risk.csv', index=False)
```
